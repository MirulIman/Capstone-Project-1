<h2> 12. Heatmap: Correlation between gender, total amount spent, quantity purchased, age and product category </h2>

```python
# Apply one-hot encoding to the "Product Category" column
df_encoded = pd.get_dummies(df, columns=["Product Category"], dtype=int)

# Verify data structure after encoding
df_encoded.head()

# Convert "Gender" to numeric values (Male=0, Female=1)
df_encoded["Gender"] = df_encoded["Gender"].map({"Male": 0, "Female": 1})

# Calculate the correlation matrix
correlation_matrix = df_encoded.corr() # This line is added to calculate the correlation matrix

# Generate heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap Including One-Hot Encoded Product Categories")
plt.show()
```

Heatmap Insights

This heatmap reveals correlations between different variables, including one-hot encoded product categories:

Gender vs. Product Categories

*  If a strong correlation exists, one gender prefers certain product categories.

Age vs. Product Categories

*  Helps identify if certain age groups favor specific product types.

Total Amount vs. Product Categories

*  A high correlation indicates which product categories contribute the most to revenue.

<h2>Gender Based Seosonal Trend</h2>

```python
# Gender-based seasonal purchasing behavior
df['month'] = df['Date'].dt.month
gender_seasonal = df.groupby(['month', 'Gender'])['Total Amount'].sum().reset_index()
plt.figure(figsize=(12,6))
sns.lineplot(data=gender_seasonal, x='month', y='Total Amount', hue='Gender', marker='o', palette={'Male': 'red', 'Female': 'blue'})
plt.title("Gender-Based Seasonal Purchasing Behavior")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()

gender_seasonal
```

```python
# Extract peak sales months for each gender
peak_sales_months_male = gender_seasonal[gender_seasonal['Gender'] == 'Male'].sort_values(by='Total Amount', ascending=False).head(2)
peak_sales_months_female = gender_seasonal[gender_seasonal['Gender'] == 'Female'].sort_values(by='Total Amount', ascending=False).head(2)

print("Peak sales months for males:")
print(peak_sales_months_male)

print("Peak sales months for females:")
print(peak_sales_months_female)
```
